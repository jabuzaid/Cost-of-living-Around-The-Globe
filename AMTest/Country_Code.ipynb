```python
import pandas as pd
```

```python
country_df = pd.read_csv("https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv")
```

```python
country_df.head()
```

```python
data = pd.read_csv("cost_of_living_data_combine.csv")
```

```python
data.head()
```

```python
df = pd.merge(data,country_df[["name","alpha-2"]],how="left",left_on="Country",right_on = "name")
```

```python
result = df.groupby("alpha-2").mean()[["CostofLivingIndex","LocalPurchasingPowerIndex"]].reset_index()
```

```python
result = result.rename(columns={"alpha-2":"id"})
```

```python
result.to_json("data.json",orient="records")
```
